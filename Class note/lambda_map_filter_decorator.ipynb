### Lambda Function

A lambda function is a small, anonymous (unnamed) function defined using the lambda keyword.

### syntax

#### lambda arguments: expression

- Its can take any numbers of arguments
- It can have only one expression 
- It automaticallly return the result of the expresssion 
- No retrun keyword required 

```python
def add(a,b):
    return a+b 
```

```python
addition=add(2,3)
```

```python
addition
```

```python
addition=lambda a,b: a+b
addition(4,8)
```

```python
## lambda  with single argument 
square =lambda x: x*x
square(9)
```

```python
## lambda with  multiple arguments
area = lambda l,w:l*w
area (4,5)
```

```python
#lambda with conditional statement
is_even= lambda sample: "even "if sample %2 ==0 else "odd"
is_even
```

```python
def to_upper(s):
    return s.upper()

to_upper("knjndnfjnfvnfnvjrvfnvifjvfuv")
```

```python
# same using lambda to convert the uppercase 
to_upper=lambda sample:sample.upper()
to_upper("knjnfvnvv dvfbhifjgunffkjkvf")
```

```python
# this is the normal way to rev the string
def rev_string(s):
    return s[::-1]

rev_string("love")
```

```python
# creating the reverse the string using the lambda
reverse_string=lambda s :s[::-1]
reverse_string("RAM")
```

### what is map()? 

- applies a fucniton to every element 
- returns a map object (convert to list if needed)

```python
# map()
celcius=(0,10, 20, 30, 40)
fahrenhiet=list(map(lambda c:(c*9/5)+32,celcius))
fahrenhiet
```

```python
list[fahrenhiet]
```

```python
## syntax for map 

# list(map(lambda x: x, sequence))
```

```python
## map(function, iterable)
```

```python
# Square all number in list  using map
num=[1,2,3,4,5]
result=list(map(lambda s:s**2, num))
result
```

### filter()

filter(function, iterable)

It filters elements from an iterable (like list, tuple).

Keeps only those elements for which the function returns True.

Returns a filter object (convert to list if needed).

```python
## map() is used to transform data 
## filter() select data


# syntax 

# filter(function, iterable)
```

```python
num=[1,2,3,4,5]
res=list(filter(lambda x: x%2==0,num))
res
```

```python
## filetr strigns longer than 4 chars 

words = ["data", "science", "ai", "machine", "computer", "cat", "dog"]

result=list(filter(lambda s: len(s)> 4,words))
result
```

```python
##Decorators
#take another Function as input 
# Add extra Functionality
#return a modified fucntion
#without chnageing the original fucntion code 
```

```python
# this is the original function
def greet():
    print("Hello!, how are you ?")
```

```python
greet()
```

```python
# basic Decorator structure 
def decorator_function(orginal_function):
    
    def wrapper():
        print("Before fucntion runs")
        orginal_function()
        print("After function runs")
        
    return wrapper   
```

```python
decorated = decorator_function(greet)
```

```python
decorated()
```
