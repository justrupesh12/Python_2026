- % = Modulus (Remainder) 10% 4

## Comparison (relational) operators

- Used to compare two values.
- They return True or False

- '==' : Equal to
- '!=' : Not equal to
- '>' : Greater than
- '<' : Less than
- '>=' : Greater than or equal to
- '<=' : Less than or equal to

```python
print(5 == 3)    # False
print(5 == 5)    # True
print(5 != 5)    # False
```

```python
x = 10
y = 20
```

```python
print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
```

## Logical Opertors

- Used to combine multiple conditions
- These operators also return True or False

opertors:
1. and : True if both conditions are True (5 > 3 and 10 > 5)

2. or : True is any one condition is True (5 > 3 or 10 < 5)

3. not : Reverse the result
not(5 > 3)



```python
(5 > 3 and 10 > 5)    # both the condition should be true to get true as output
```

```python
(5 > 3 and 10 < 5)
```

```python
(5 > 3 or 10 < 5)   # at least one shold be True to get True as output
```

```python
(5 > 3 or 10 > 5)
```

```python
(5 < 3 or 10 < 5)
```

```python
not(5 > 3)
```

```python
not(5 < 3)
```
