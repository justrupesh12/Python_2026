```python
# Write a program to count the vowel in the string
Vowel="aeiou"
name ="Rupesh"
count=0
for i in Vowel:
    if i in name.lower():
        count=count+1
print(count)        
        
        
        
```

```python
# check weather String is palindram or not 

name=input("Enter the name ")
reverse=name[::-1]
if name== reverse:
    print("this is palindram")
else:
    print("This is not palindram")    
```
