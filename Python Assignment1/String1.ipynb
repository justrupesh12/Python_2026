```python
#1.	Given a string, remove leading and trailing spaces.

hi= "   Hello, World!   "
print(hi.strip())
```

```python
#2.	Check whether a string starts with a given character.
hi= "Hello, World!   "
print(hi.startswith("H"))
```

```python
#3.	Find the index of the first occurrence of a character in a string.
hi= "Hello, World!"
print(hi.find("o"))
print(hi.find("z"))  # (-1 means character not found)
```

```python
#4.	Count how many times a specific character appears in a string.

hi="I love programming in Python. Python is great!"
print(hi.count("o"))
print(hi.count("P"))
print(hi.count("z"))  # (0 means character not found)
```

```python
#5.	Split a sentence into a list of words.
hi="I love programming in Python. Python is great!"
print(hi.split())
```

```python
#6.	Join a list of words into a single string with spaces.
hi_list = ["I", "love", "programming", "in", "Python."]
hi_string = " ".join(hi_list)
print(hi_string)
```

```python
#7.	Replace a word in a sentence with another word
hi_list = ["I", "love", "programming", "in", "Python."]

hi_list[0]="you"
print(hi_list)
```

```python
#7.	Replace a word in a sentence with another word
hi="I love programming in Python. Python is great!"

r=hi.replace("I","You")
print(r)
```
