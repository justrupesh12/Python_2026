```python
# Question 34: Count the depth of nesting in [1, [2, 3], [4, [5, 6]], 7]
print("\nQuestion 34: Count the depth of nesting in [1, [2, 3], [4, [5, 6]], 7]")
# Your code here

l=  [1, [2, 3], [4, [5, 6]], 7]
for i in l:
    if type(i)==list:
        for j in i:
            if type(j)==list:
                for k in j:
                    if type(k)==list:
                        print("Depth is 4")
                    else:
                        print("Depth is 3")
            else:
                print("Depth is 2")
    else:
        print("Depth is 1")
print("Maximum depth is 4")        
```
