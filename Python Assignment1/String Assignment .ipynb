```python
# Question 41: Check if string contains balanced parentheses: "((()))"
print("\nQuestion 41: Check if string contains balanced parentheses: '((())))'")
# Your code here
s='(((())))))))'
count=0
for i in s:
    if i == "(":
        count += 1
    elif i == ")":
        count -= 1
        if count < 0:
            break

if count == 0:
    print("Balanced parentheses")
else:
    print("Unbalanced parentheses")
          
```

```python
# Question 43: Find the longest common substring between "programming" and "grammar"
print("\nQuestion 43: Find the longest common substring between 'programming' and 'grammar'")
# Your code here
s='programming' 
m='grammer'
long=""
for i in range(len(s)):
    for j in range(i+1,len(s)+1):
        sub = s[i:j]
        if sub in m and len(sub)>len(long):
            long=sub
print(long)        
```

```python
print("\nQuestion 43: Find the longest common substring between 'programming' and 'grammar'")

s = "programming"
m = "grammar"

longest = ""

# Try all substrings of s
for i in range(len(s)):
    for j in range(i+1, len(s)+1):
        sub = s[i:j]
        if sub in m and len(sub) > len(longest):
            longest = sub

print("Longest Common Substring:", longest)
```

```python
# Another way
def rot13(c):
    if 'a' <= c <= 'z':
        return chr((ord(c) - ord('a') + 13) % 26 + ord('a'))
    elif 'A' <= c <= 'Z':
        return chr((ord(c) - ord('A') + 13) % 26 + ord('A'))
    else:
        return c
    
print(rot13('h'))  # Outputs: u
```
